```python
import matplotlib.pyplot as plt

# Import pandas for data manipulation and analysis
import pandas as pd

# Import numpy for numerical computing
import numpy as np

import seaborn as sns
```

```python
df_risk_management = pd.read_csv("german_credit_data.csv")
df_risk_management.describe()
```

```python
df_risk_management.head()
```

```python
df_risk_management.dtypes
```

```python
#This shows something
# Create a count plot
plt.figure(figsize=(10,6))
sns.countplot(x='Job', hue='Risk', data=df_risk_management)

# Set labels
plt.xlabel("Job Type")
plt.ylabel("Count")
plt.title("Count plot separated by job type and risk")

# Show the plot
plt.show()
```

# Credit Risk by Job Type

The bar chart above represents the distribution of credit risk ('good' vs 'bad') across different job types (labeled 0 to 3). Key observations from the chart are:

- **Job Type 2**: Shows the highest count, indicating it's the most common job category in the dataset. A significant majority of individuals in this category are considered a 'good' credit risk, which could imply that job type 2 is viewed as more financially stable or reliable.
- **Job Type 1**: Although fewer in number, individuals in this job category have a larger proportion of 'bad' risk, hinting at potential financial instability or higher risk in this employment sector.
- **Job Types 0 and 3**: These categories have the lowest counts, suggesting they are less common within the population. Both categories show a relatively balanced distribution of 'good' and 'bad' credit risks.

This analysis can provide valuable insights for financial institutions to tailor their risk assessment models according to job type, potentially improving the accuracy of credit scoring algorithms.

```python
# Create separate dataframes for each category
df_good = df_risk_management[df_risk_management['Risk'] == 'good']
df_bad = df_risk_management[df_risk_management['Risk'] == 'bad']

# Define the range of the bins
binrange = (df_risk_management["Credit amount"].min(), df_risk_management["Credit amount"].max())

# Create histograms
sns.histplot(data=df_good, x="Credit amount", color='blue', binwidth=1000, alpha=0.5, edgecolor="black", linewidth=1, kde=True, binrange=binrange, label='Good')
sns.histplot(data=df_bad, x="Credit amount", color='red', binwidth=1000, alpha=0.8, edgecolor="black", linewidth=1, kde=True, binrange=binrange, label='Bad')

# Add labels and title
plt.xlabel("Credit amount")
plt.ylabel("Count")
plt.title("Histogram of Credit amount by Credit Risk")

# Add a legend
plt.legend(title="Credit Risk")

# Show the plot
plt.show()
```

# Credit Amount by Credit Risk

The histogram displays the distribution of credit amounts for loans, categorized by credit risk ('Good' vs 'Bad').

Observations:
- **Predominance of Smaller Loans**: Both 'Good' and 'Bad' credit risks are most commonly associated with smaller loans, as indicated by the higher counts in lower credit amount bins.
- **Distribution Spread**: 'Good' risk loans show a broader spread towards higher credit amounts, suggesting these individuals are often approved for larger loans.
- **Ratio Balance at Higher Amounts**: Notably, the ratio of 'Good' to 'Bad' risk loans tends to even out as the credit amount increases. This trend implies a more nuanced risk profile for larger loans and highlights the need for careful credit risk assessment in higher loan categories.

This visualization highlights the relationship between the amount of credit and the associated risk classification.


```python
# Define the range of the bins
binrange = (df_risk_management["Duration"].min(), df_risk_management["Duration"].max())

# Create histograms
sns.histplot(data=df_good, x="Duration", color='blue', binwidth=5, alpha=0.5, edgecolor="black", linewidth=1, kde=True, binrange=binrange, label='Good')
sns.histplot(data=df_bad, x="Duration", color='red', binwidth=5, alpha=0.8, edgecolor="black", linewidth=1, kde=True, binrange=binrange, label='Bad')

# Add labels and title
plt.xlabel("Duration")
plt.ylabel("Count")
plt.title("Histogram of Duration by Loan Status")

# Add a legend
plt.legend(title="Credit Risk")

# Show the plot
plt.show()
```
